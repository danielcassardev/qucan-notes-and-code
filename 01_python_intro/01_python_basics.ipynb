# Git

#### git add
Staging the commit i.e. these are the changes I want to include in the next commit.
using the suffix ```.``` means everything within this folder and sub-folders
else include the specific file path and name ```01_python_intro\01_python_basics.ipynb```
Example: ```git add . ```

#### git commit
Save progress and describe what was performed or changed in natural language
Example: ```git commit -m "message"```

#### git push
Uploads commits to GitHub. Everything only exists locally until this is performed
Example: ```git push```

# Formatted Strings
Different methods of presenting text

````python
#The rudimentary and inefficiency method, noting age has to be stored as a string
name = "Daniel"
age = "20"
print (str("My name is " + name + " and I am " + age + " years old"))
#Cleaner method (f-string) which evaluates anything within {} as a string
name = "Daniel"
age = 20
print(f"My name is {name} and I am {age} years old")
````

# Data Types

# Integer Math Operations

````python
x = 10
y = 3
print("Addition", x + y)
print("Subtraction", x - y)
print("Multiplication", x * y)
print("Division", x / y)
print("Integer division", x // y)
print("Modulus (remainder)", x % y)
print("Exponent", x ** y)
````

````python
x, y = y, x
print("The new x value", x)
print("The new y value", y)
````

# Floats

````python
#Using floats to find the area of a circle
pi = 3.14149
radius = 4
area = pi * radius **2
print(radius**2)
print(area)
````

````python
#Using floats to calculate Energy value
#E = hf
#Where h = Planks constant and f = frequency
h = 1.16
f = 30
E = h*f
print(E)
````
