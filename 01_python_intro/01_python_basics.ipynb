# Git

#### git add
Staging the commit i.e. these are the changes I want to include in the next commit.
using the suffix ```.``` means everything within this folder and sub-folders
else include the specific file path and name ```01_python_intro\01_python_basics.ipynb```
Example: ```git add . ```

#### git commit
Save progress and describe what was performed or changed in natural language
Example: ```git commit -m "message"```

#### git push
Uploads commits to GitHub. Everything only exists locally until this is performed
Example: ```git push```

````python
numbers =[4,1,3,2,5]
print(numbers)
numbers.sort()
print(numbers)
numbers.reverse()
print(numbers)
numbers.reverse()
#single colon : is to slice from A to B [start : end]
print(numbers[:3])
#with no start value, python assumes 0
print(numbers[1:3])
#starts at the second value

#double colon :: is to slice with a step [start : step : end]
print(numbers[::-1])
#with no start or step, the whole sequence is reversed
````

# Classes and Objects

````python
class Dog:
#This class defines the blueprint for creating Dog objects

    def bark(self):
        #Method that belongs to the Dog class; 'self' is the instance calling it
        print("Woof Woof")
    def greet(self):
        #Another method of the Dog instance
        print("lick")

my_dog = Dog()      #create an instance of the dog class
my_dog.bark()       #call the bark method on this instance
my_dog.greet()      #call the greet method on this instance

#Classes describe what an object *is* and *can do*
#Objects are actual examples created from the class blueprint
````

````python
class Dog:
    def __init__(self, name, breed, age):
            #__init__ initializes the object's attributes when it's created
           self.name = name
           self.breed = breed
           self.age = age

    def bark(self):
        #Instance method â€” 'self' is the specific object calling the method
        print("Woof Woof")
    def greet(self):
        print("lick")

my_dog = Dog("Drake", "Husky", 6)
my_dog2 = Dog("Dewy", "Golden Retriever", 3)
print(f"{my_dog.name} is a {my_dog.breed} and is {my_dog.age} years old")
print(f"{my_dog2.name} is a {my_dog2.breed} and is {my_dog2.age} years old")
my_dog.bark()
````
